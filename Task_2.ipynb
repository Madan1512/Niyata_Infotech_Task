```python
import numpy as np
data["next_week_purchase"]=np.where(data["Frequency of Purchases"].str.lower()=="weekly",1,0)
```

```python
features = ['Age', 'Gender', 'Location', 'Purchase Amount (USD)', 'Previous Purchases', 'Subscription Status']
x= data[features]
y = data['next_week_purchase']
```

```python
x.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
categorical=["Gender","Location","Subscription Status"]
for col in categorical:
  x[col]=encoder.fit_transform(x[col])
```

```python
x.info()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
numerical=["Age","Purchase Amount (USD)","Subscription Status"]
x_train[numerical]=scaler.fit_transform(x_train[numerical])
x_test[numerical]=scaler.transform(x_test[numerical])
```
